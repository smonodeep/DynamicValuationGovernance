```python
import re
import pandas as pd

with open("/content/reseach_plan_cleaned.bib", "r", encoding="utf-8") as f:
    text = f.read()

# Split entries manually
entries = re.split(r'(?=@\w+\{)', text)

entries = [e.strip() for e in entries if e.strip()]

print("Recovered Entries:", len(entries))

records = []

for entry in entries:

    def extract_field(field):
        pattern = rf'{field}\s*=\s*[\{{"](.+?)[\}}"],?\n'
        match = re.search(pattern, entry, re.DOTALL | re.IGNORECASE)
        return match.group(1).replace("\n", " ").strip() if match else ""

    title = extract_field("title")
    author = extract_field("author")
    year = extract_field("year")
    doi = extract_field("doi")
    journal = extract_field("journal")
    abstract = extract_field("abstract")

    records.append({
        "Title": title,
        "Author": author,
        "Year": year,
        "DOI": doi,
        "Journal": journal,
        "Abstract": abstract
    })

df = pd.DataFrame(records)

print(df.head())
print("Final Parsed Rows:", len(df))
```

```python
import pandas as pd

# Assuming 'df' is the DataFrame from previous steps

new_df_records = []

for index, row in df.iterrows():
    doi = row['DOI'] if pd.notna(row['DOI']) and row['DOI'] != '' else "Could not find details"
    publication = row['Journal'] if pd.notna(row['Journal']) and row['Journal'] != '' else "Could not find details"
    title = row['Title'] if pd.notna(row['Title']) and row['Title'] != '' else "Could not find details"
    authors = row['Author'] if pd.notna(row['Author']) and row['Author'] != '' else "Could not find details"
    year = row['Year'] if pd.notna(row['Year']) and row['Year'] != '' else "Could not find details"
    abstract = row['Abstract'] if pd.notna(row['Abstract']) and row['Abstract'] != '' else "Could not find details"

    citation_url = "Could not find details"
    if doi != "Could not find details":
        citation_url = f"https://doi.org/{doi}"

    new_df_records.append({
        'S.No': index + 1,
        'DOI': doi,
        'Publication': publication,
        'Title': title,
        'Subtitle': "Could not find details", # Not directly available, per instruction
        'Authors': authors,
        'Year': year,
        'Citation / URL': citation_url,
        'Exact Abstract': abstract,
        'Key Findings': "Could not find details" # Not directly available, per instruction
    })

final_df = pd.DataFrame(new_df_records)

print("New DataFrame created with the specified columns and missing value handling:")
print(final_df.head())
print(f"Total entries in the new DataFrame: {len(final_df)}")
```

```python
import pandas as pd

def create_markdown_table_from_bib(bib_df, filename='output_table.md'):
    """
    Creates a markdown table from bib_df with specific columns and saves it to a file.
    """
    output_rows = []

    for index, row in bib_df.iterrows():
        doi = row.get('doi', '')
        publication = row.get('journal', row.get('publisher', '')) # Prioritize journal, then publisher
        title = row.get('title', '')
        subtitle = '' # Not directly available in bib_df
        authors = row.get('author', '')
        year = row.get('year', '')

        citation_url = ''
        if pd.notna(row.get('url')) and row.get('url') != '':
            citation_url = row.get('url')
        elif pd.notna(doi) and doi != '':
            citation_url = f"https://doi.org/{doi}"

        abstract = row.get('abstract', '')
        key_findings = '' # Not directly available in bib_df

        output_rows.append({
            'S.No': index + 1,
            'DOI': doi,
            'Publication': publication,
            'Title': title,
            'Subtitle': subtitle,
            'Authors': authors,
            'Year': year,
            'Citation / URL': citation_url,
            'Exact Abstract': abstract,
            'Key Findings': key_findings
        })

    output_df = pd.DataFrame(output_rows)

    # Convert DataFrame to markdown table string, including header and separator
    markdown_table_string = output_df.to_markdown(index=False)

    with open(filename, 'w', encoding='utf-8') as f:
        f.write(markdown_table_string)

    print(f"Markdown table saved to {filename}")

# Assuming bib_df is already loaded and available from previous steps
create_markdown_table_from_bib(bib_df, filename='research_plan_summary.md')

# Display the content of the created file for verification
with open('research_plan_summary.md', 'r', encoding='utf-8') as f:
    print(f.read())
```
